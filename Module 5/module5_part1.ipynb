# For Loop in Python

## Introduction
The `for` loop in Python is used to iterate over a sequence (such as a list, tuple, dictionary, string, or range) and execute a block of code multiple times.

## Syntax
```python
for variable in sequence:
    # Code to execute
```

## Example 1: Looping Through a List
```python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

## Example 2: Using `range()`
```python
for i in range(5):
    print(i)  # Prints numbers from 0 to 4
```

## Example 3: Iterating Over a String
```python
for char in "Python":
    print(char)
```

## Example 4: Using `break`
```python
for num in range(10):
    if num == 5:
        break  # Stops loop when num is 5
    print(num)
```

## Example 5: Using `continue`
```python
for num in range(5):
    if num == 2:
        continue  # Skips iteration when num is 2
    print(num)
```

## Example 6: Nested For Loop
```python
for i in range(3):
    for j in range(2):
        print(f"i={i}, j={j}")
```

## Conclusion
The `for` loop is a powerful construct that simplifies iteration in Python. It is commonly used for traversing sequences, performing repetitive tasks, and handling collections efficiently.



# For Loop Exercises in Python

## Exercise 1: Print Numbers from 1 to 10
**Task:** Write a Python program using a `for` loop to print numbers from 1 to 10.
```python
for i in range(1, 11):
    print(i)
```

## Exercise 2: Sum of First N Natural Numbers
**Task:** Write a program to calculate the sum of the first N natural numbers.
```python
var = 0

for i in range(1,5):
    var = var + i

print(var)
```

## Exercise 3: Print Even Numbers from 1 to 20
**Task:** Use a `for` loop to print all even numbers from 1 to 20.
```python
for i in range(2, 21, 2):
    print(i)
```

## Exercise 4: Reverse a String
**Task:** Write a Python program to reverse a string using a `for` loop.
```python
a = ""
b = "python"

for i in b:
    a = i + a
    
print(a)
```

## Exercise 5: Multiplication Table
**Task:** Write a program to print the multiplication table of a given number.
```python
num = int(input("Enter a number: "))
for i in range(1, 11):
    print(f"{num} x {i} = {num * i}")
```

## Exercise 6: Count Vowels in a String
**Task:** Write a program to count the number of vowels in a given string.
```python
a = "aeiou"
b = "youtube"
y = 0

for i in b:
    if i in a:
        y = y + 1
    
print(y)
```

## Exercise 7: Print a Pattern
**Task:** Write a program to print the following pattern:
```
*
**
***
****
*****
```
```python
for i in range(1, 6):
    print("*" * i)
```
